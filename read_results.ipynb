```python
from __future__ import annotations
import pickle

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sstt
```

```python
# results file of each experiment
fullpath_results_exp_0 = './results/results_experiment_0.pkl'
fullpath_results_exp_1 = './results/results_experiment_1.pkl'
fullpath_results_exp_2 = './results/results_experiment_2.pkl'
fullpath_results_exp_3 = './results/results_experiment_3.pkl'

# load all results files

with open(fullpath_results_exp_0, 'rb') as f:
    results_exp_0 = pickle.load(f)['results']

with open(fullpath_results_exp_1, 'rb') as f:
    results_exp_1 = pickle.load(f)['results']

with open(fullpath_results_exp_2, 'rb') as f:
    results_exp_2 = pickle.load(f)['results']

with open(fullpath_results_exp_3, 'rb') as f:
    results_exp_3 = pickle.load(f)['results']

# check the number of repetitions
num_repetitions_exp_0 = results_exp_0['NUM_REPETITIONS']
num_repetitions_exp_1 = results_exp_1['NUM_REPETITIONS']
num_repetitions_exp_2 = results_exp_2['NUM_REPETITIONS']
num_repetitions_exp_3 = results_exp_3['NUM_REPETITIONS']
print(
    f"Repetitions of Exp 0:\t{num_repetitions_exp_0:d}\n"
    f"Repetitions of Exp 1:\t{num_repetitions_exp_1:d}\n"
    f"Repetitions of Exp 2:\t{num_repetitions_exp_2:d}\n"
    f"Repetitions of Exp 3:\t{num_repetitions_exp_3:d}\n"
)
```

```python
# Format AUROC data

# training AUROC
auroc_train_q_exp_0 = [results_exp_0['repetition'][_]['metrics_train_q']['auroc'] for _ in range(num_repetitions_exp_0)]
auroc_train_q_exp_1 = [results_exp_1['repetition'][_]['metrics_train_q']['auroc'] for _ in range(num_repetitions_exp_1)]
auroc_train_q_exp_2 = [results_exp_2['repetition'][_]['metrics_train_q']['auroc'] for _ in range(num_repetitions_exp_2)]
auroc_train_q_exp_3 = [results_exp_3['repetition'][_]['metrics_train_q']['auroc'] for _ in range(num_repetitions_exp_3)]

# validation AUROC
auroc_valid_q_exp_0 = [results_exp_0['repetition'][_]['metrics_valid_q']['auroc'] for _ in range(num_repetitions_exp_0)]
auroc_valid_q_exp_1 = [results_exp_1['repetition'][_]['metrics_valid_q']['auroc'] for _ in range(num_repetitions_exp_1)]
auroc_valid_q_exp_2 = [results_exp_2['repetition'][_]['metrics_valid_q']['auroc'] for _ in range(num_repetitions_exp_2)]
auroc_valid_q_exp_3 = [results_exp_3['repetition'][_]['metrics_valid_q']['auroc'] for _ in range(num_repetitions_exp_3)]

# validation sensitivity (maximizing balanced accuracy)
sens_valid_q_exp_0 = [results_exp_0['repetition'][_]['metrics_valid_q']['sensitivity_maxba'] for _ in range(num_repetitions_exp_0)]
sens_valid_q_exp_1 = [results_exp_1['repetition'][_]['metrics_valid_q']['sensitivity_maxba'] for _ in range(num_repetitions_exp_1)]
sens_valid_q_exp_2 = [results_exp_2['repetition'][_]['metrics_valid_q']['sensitivity_maxba'] for _ in range(num_repetitions_exp_2)]
sens_valid_q_exp_3 = [results_exp_3['repetition'][_]['metrics_valid_q']['sensitivity_maxba'] for _ in range(num_repetitions_exp_3)]

# validation specificity (maximizing balanced accuracy)
spec_valid_q_exp_0 = [results_exp_0['repetition'][_]['metrics_valid_q']['specificity_maxba'] for _ in range(num_repetitions_exp_0)]
spec_valid_q_exp_1 = [results_exp_1['repetition'][_]['metrics_valid_q']['specificity_maxba'] for _ in range(num_repetitions_exp_1)]
spec_valid_q_exp_2 = [results_exp_2['repetition'][_]['metrics_valid_q']['specificity_maxba'] for _ in range(num_repetitions_exp_2)]
spec_valid_q_exp_3 = [results_exp_3['repetition'][_]['metrics_valid_q']['specificity_maxba'] for _ in range(num_repetitions_exp_3)]

# validation balanced accuracy (maximizing balanced accuracy)
ba_valid_q_exp_0 = [results_exp_0['repetition'][_]['metrics_valid_q']['balanced_accuracy_maxba'] for _ in range(num_repetitions_exp_0)]
ba_valid_q_exp_1 = [results_exp_1['repetition'][_]['metrics_valid_q']['balanced_accuracy_maxba'] for _ in range(num_repetitions_exp_1)]
ba_valid_q_exp_2 = [results_exp_2['repetition'][_]['metrics_valid_q']['balanced_accuracy_maxba'] for _ in range(num_repetitions_exp_2)]
ba_valid_q_exp_3 = [results_exp_3['repetition'][_]['metrics_valid_q']['balanced_accuracy_maxba'] for _ in range(num_repetitions_exp_3)]

# convert all to NumPy arrays

auroc_train_q_exp_0 = np.array(auroc_train_q_exp_0, dtype=np.float32)
auroc_train_q_exp_1 = np.array(auroc_train_q_exp_1, dtype=np.float32)
auroc_train_q_exp_2 = np.array(auroc_train_q_exp_2, dtype=np.float32)
auroc_train_q_exp_3 = np.array(auroc_train_q_exp_3, dtype=np.float32)

auroc_valid_q_exp_0 = np.array(auroc_valid_q_exp_0, dtype=np.float32)
auroc_valid_q_exp_1 = np.array(auroc_valid_q_exp_1, dtype=np.float32)
auroc_valid_q_exp_2 = np.array(auroc_valid_q_exp_2, dtype=np.float32)
auroc_valid_q_exp_3 = np.array(auroc_valid_q_exp_3, dtype=np.float32)

sens_valid_q_exp_0 = np.array(sens_valid_q_exp_0, dtype=np.float32)
sens_valid_q_exp_1 = np.array(sens_valid_q_exp_1, dtype=np.float32)
sens_valid_q_exp_2 = np.array(sens_valid_q_exp_2, dtype=np.float32)
sens_valid_q_exp_3 = np.array(sens_valid_q_exp_3, dtype=np.float32)

spec_valid_q_exp_0 = np.array(spec_valid_q_exp_0, dtype=np.float32)
spec_valid_q_exp_1 = np.array(spec_valid_q_exp_1, dtype=np.float32)
spec_valid_q_exp_2 = np.array(spec_valid_q_exp_2, dtype=np.float32)
spec_valid_q_exp_3 = np.array(spec_valid_q_exp_3, dtype=np.float32)

ba_valid_q_exp_0 = np.array(ba_valid_q_exp_0, dtype=np.float32)
ba_valid_q_exp_1 = np.array(ba_valid_q_exp_1, dtype=np.float32)
ba_valid_q_exp_2 = np.array(ba_valid_q_exp_2, dtype=np.float32)
ba_valid_q_exp_3 = np.array(ba_valid_q_exp_3, dtype=np.float32)
```

```python
# Discard bad trainings

auroc_train_q_allexps = np.concatenate(
    (
        auroc_train_q_exp_0,
        auroc_train_q_exp_1,
        auroc_train_q_exp_2,
        auroc_train_q_exp_3,
    ),
    axis=0,
)

plt.figure(figsize=(20.0, 4.0))
plt.title("Distribution of training AUROC of all four experiments")
plt.xlabel("Training AUROC (dimensionless)")
plt.ylabel("Counts (dimensionless natural)")
bins = np.arange(start=0.0, stop=1.0, step=0.0025)
plt.hist(auroc_train_q_allexps, bins=bins, density=False, histtype='bar', color='k')
# plt.hist(auroc_train_q_exp_0, bins=bins, density=True, histtype='bar', alpha=0.5, label='Experiment 0')
# plt.hist(auroc_train_q_exp_1, bins=bins, density=True, histtype='bar', alpha=0.5, label='Experiment 1')
# plt.hist(auroc_train_q_exp_2, bins=bins, density=True, histtype='bar', alpha=0.5, label='Experiment 2')
# plt.hist(auroc_train_q_exp_3, bins=bins, density=True, histtype='bar', alpha=0.5, label='Experiment 3')
# plt.legend()
plt.grid()
xmin = auroc_train_q_allexps.min() - 0.05
xmin = max(0.0, xmin)
plt.axis([xmin, 1.0, 0.0, None])
plt.show()
```

```python
HOLE_LOCATION_BY_EYE = 0.7  # set manually by inspection of plot above

# Information on the training AUROC hole

mask_below_hole = auroc_train_q_allexps < HOLE_LOCATION_BY_EYE
mask_above_hole = auroc_train_q_allexps > HOLE_LOCATION_BY_EYE
AUROC_TRAIN_HOLE_LO = auroc_train_q_allexps[mask_below_hole].max()
AUROC_TRAIN_HOLE_HI = auroc_train_q_allexps[mask_above_hole].min()

print(
    f"Bottom of hole in distribution of training AUROC:\t{AUROC_TRAIN_HOLE_LO:.4f}\n"
    f"Top of hole in distribution of training AUROC:\t\t{AUROC_TRAIN_HOLE_HI:.4f}\n"
)

# Discard bad trainings

mask_exp_0 = auroc_train_q_exp_0 > HOLE_LOCATION_BY_EYE
mask_exp_1 = auroc_train_q_exp_1 > HOLE_LOCATION_BY_EYE
mask_exp_2 = auroc_train_q_exp_2 > HOLE_LOCATION_BY_EYE
mask_exp_3 = auroc_train_q_exp_3 > HOLE_LOCATION_BY_EYE

num_discarded_exp_0 = num_repetitions_exp_0 - mask_exp_0.sum()
num_discarded_exp_1 = num_repetitions_exp_1 - mask_exp_1.sum()
num_discarded_exp_2 = num_repetitions_exp_2 - mask_exp_2.sum()
num_discarded_exp_3 = num_repetitions_exp_3 - mask_exp_3.sum()

auroc_valid_q_exp_0 = auroc_valid_q_exp_0[mask_exp_0]
auroc_valid_q_exp_1 = auroc_valid_q_exp_1[mask_exp_1]
auroc_valid_q_exp_2 = auroc_valid_q_exp_2[mask_exp_2]
auroc_valid_q_exp_3 = auroc_valid_q_exp_3[mask_exp_3]

sens_valid_q_exp_0 = sens_valid_q_exp_0[mask_exp_0]
sens_valid_q_exp_1 = sens_valid_q_exp_1[mask_exp_1]
sens_valid_q_exp_2 = sens_valid_q_exp_2[mask_exp_2]
sens_valid_q_exp_3 = sens_valid_q_exp_3[mask_exp_3]

spec_valid_q_exp_0 = spec_valid_q_exp_0[mask_exp_0]
spec_valid_q_exp_1 = spec_valid_q_exp_1[mask_exp_1]
spec_valid_q_exp_2 = spec_valid_q_exp_2[mask_exp_2]
spec_valid_q_exp_3 = spec_valid_q_exp_3[mask_exp_3]

ba_valid_q_exp_0 = ba_valid_q_exp_0[mask_exp_0]
ba_valid_q_exp_1 = ba_valid_q_exp_1[mask_exp_1]
ba_valid_q_exp_2 = ba_valid_q_exp_2[mask_exp_2]
ba_valid_q_exp_3 = ba_valid_q_exp_3[mask_exp_3]

print(
    f"Experiment 0: discarded {num_discarded_exp_0:d} training repetitions out of {num_repetitions_exp_0:d}\n"
    f"Experiment 1: discarded {num_discarded_exp_1:d} training repetitions out of {num_repetitions_exp_1:d}\n"
    f"Experiment 2: discarded {num_discarded_exp_2:d} training repetitions out of {num_repetitions_exp_2:d}\n"
    f"Experiment 3: discarded {num_discarded_exp_3:d} training repetitions out of {num_repetitions_exp_3:d}\n"
)
```

```python
def location_tests(x, y):
    """
    Performs statistical tests for difference in distribution location:
    - Kolmogorov-Smirnov
    - Mood's median
    - SciPy's "rank-sums"
    - SciPy's "Mann-Whitney"
    All tests are non-parametric, 2-samples, unpaired, and one-tailed.
    """

    # Naming convention for p-value variables:
    # - "pks" for Kolmogorov-Smirnov
    # - "pmm" for Mood's median
    # - "prs" for SciPy's "ranksums"
    # - "pmw" for SciPy's "Mann-Whitney"

    # Kolmogorov-Smirnov test
    # NB: only one with alternative reversed, due to SciPy's implementation
    _, pks = sstt.ks_2samp(x, y, alternative='less', method='auto')

    # Mood's median test
    _, pmm, _, _ = sstt.median_test(x, y, ties='ignore', nan_policy='raise')

    # SciPy's "rank-sums" test
    _, prs = sstt.ranksums(x, y, alternative='greater')

    # SciPy's "Mann-Whitney" test
    pmw = sstt.mannwhitneyu(x, y, alternative='greater', nan_policy='raise').pvalue

    # display the outcomes
    print(
        f"\n"
        f"Kolmogorov-Smirnov:\t{pks:.3e}\n"
        f"Mood's median:\t\t{pmm:.3e}\n"
        f"SciPy's \"rank-sums\":\t{prs:.3e}\n"
        f"SciPy's \"Mann-Whitney\":\t{pmw:.3e}\n"
        f"\n"
    )

    return


location_tests(auroc_valid_q_exp_3, auroc_valid_q_exp_2)
```
