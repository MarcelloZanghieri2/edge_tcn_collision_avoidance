```python
import numpy as np
import matplotlib.pyplot as plt

from ultras.dataset import constants as dsc
from ultras.dataset import loading as dsload
```

```python
# Example window of noiseless positive
IDX_COLL_NOISELESS_POS = 0
IDX_RUN_NOISELESS_POS = 10
IDX_WIN_NOISELESS_POS = 0

# Example window of noisy negative
IDX_COLL_NOISY_NEG = 7
IDX_RUN_NOISY_NEG = 50
IDX_WIN_NOISY_NEG = 0

s2ms = 1000.0
TIME_MS = np.arange(dsc.NUM_SAMPLES) / dsc.FS_HZ * s2ms
#MAX_TIME_MS = TIME_MS[-1]


x_noiseless_pos = dsload.load_window(
    IDX_COLL_NOISELESS_POS, IDX_RUN_NOISELESS_POS, IDX_WIN_NOISELESS_POS)

x_noisy_neg = dsload.load_window(
    IDX_COLL_NOISY_NEG, IDX_RUN_NOISY_NEG, IDX_WIN_NOISY_NEG)
```

```python
# Plot noiseless positive

fig, axs = plt.subplots(
    nrows=3, ncols=3, sharex=True, sharey=True, figsize=(16.0, 8.0))

for idx_row in range(3):
    for idx_col in range(3):

        idx_ch = idx_row * 3 + idx_col
        ax = axs[idx_row, idx_col]
        ax.set_title(f"channel {idx_ch + 1}")
        ax.plot(TIME_MS, x_noiseless_pos[idx_ch], 'b-', linewidth=2.5)
        ax.grid()
        ax.set_xlim([0.0, 20.0])
        ax.set_ylim([0.0, 255.0])

        if idx_row == 2:
            ax.set_xlabel("Time (ms)", fontsize=14)
        if idx_col == 0:
            ax.set_ylabel("Ultrasound signal\n(8-bit ADC counts)", fontsize=14)

plt.tight_layout()
plt.savefig("./plot_noiseless_pos.pdf", dpi=600, bbox_inches='tight')
plt.show()


# Plot noisy negative

fig, axs = plt.subplots(
    nrows=3, ncols=3, sharex=True, sharey=True, figsize=(16.0, 8.0))

for idx_row in range(3):
    for idx_col in range(3):
        
        idx_ch = idx_row * 3 + idx_col
        ax = axs[idx_row, idx_col]
        ax.set_title(f"channel {idx_ch + 1}")
        ax.plot(TIME_MS, x_noisy_neg[idx_ch], 'b-', linewidth=2.5)
        ax.grid()
        ax.set_xlim([0.0, 20.0])
        ax.set_ylim([0.0, 255.0])

        if idx_row == 2:
            ax.set_xlabel("Time (ms)", fontsize=14)
        if idx_col == 0:
            ax.set_ylabel("Ultrasound signal\n(8-bit ADC counts)", fontsize=14)

plt.tight_layout()
plt.savefig("./plot_noisy_neg.pdf", dpi=600, bbox_inches='tight')
plt.show()
```
